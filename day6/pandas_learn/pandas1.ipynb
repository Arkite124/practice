```python
from pandas import Series, DataFrame

daeshin = {
    'open':  [11650, 11100, 11200, 11100, 11000],
    'high':  [12100, 11800, 11200, 11100, 11150],
    'low':   [11600, 11050, 10900, 10950, 10900],
    'close': [11900, 11600, 11000, 11100, 11050]
}
date=['16.02.29','16.02.26','16.02.25','16.02.24','16.02.23']
daeshin_day = DataFrame(daeshin,index=date)
print(daeshin_day)
# dataframe은 행 제목 먼저, 열 제목 나중에
# print("open 16.02.29 :",daeshin_day['open']['16.02.29'])
# loc과 iloc의 차이
# loc은 이름 기준, iloc은 열 순서 기준
print("daeshin_day.loc['16.02.29':'16.02.25'] :",daeshin_day.loc['16.02.29':'16.02.25'])
print("daeshin_day.iloc[0:2] :",daeshin_day.iloc[0:2])
```
