```python
import pandas as pd

# 데이터 정의
df = pd.DataFrame({
    '2007': [7.71, 19.62, 10.47, 10.87, 4.04, 2.01],
    '2008': [7.95, 17.71, 8.45, 10.83, 3.78, 2.05],
    '2009': [11.06, 15.00, 5.58, 7.55, 3.45, 1.50],
    '2010': [15.84, 16.70, 7.60, 9.09, 4.20, 2.25],
    '2011': [16.33, 17.48, 8.40, 7.88, 4.62, 2.54]
}, index=(['China', 'EU', 'US', 'Japan', 'Korea', 'Mexico']))
df.loc['total'] = df.sum(axis=0)
df.loc['mean'] = df.mean(axis=0)
print(df)
```

```python
import pandas as pd

# 데이터프레임 생성
d_df = pd.DataFrame(data = [[10, 20, 30, 40], [50, 60, 70, 80]],
                    columns = ['A', 'B', 'C', 'D'])
# 'B' 열을 삭제한 새로운 데이터프레임 생성
new_df = d_df.drop('B', axis=1, inplace=False)
# 결과 출력
print(d_df)
print(new_df)
```
