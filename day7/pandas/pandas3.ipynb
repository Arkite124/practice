```python
import pandas as pd
# CSV 파일이 저장된 기본 경로 설정
path = 'https://github.com/dongupak/DataML/raw/main/csv/'
weather_file = path + 'weather.csv'
# encoding='CP949' 옵션을 사용하여 한글 데이터를 포함한 csv 파일을 읽어옴
weather = pd.read_csv(weather_file, index_col = 0, encoding='CP949')
# 데이터프레임의 상위 3개 행만 출력
print(weather.head(3))
print('weather 데이터의 shape :', weather.shape)
print(weather.describe())
print('weather 데이터의 평균 :', weather.mean())
print('weather 데이터의 표준편차 :', weather.std())
```

```python
missing_data=weather[weather['평균풍속'].isna()]
print(missing_data)
```

```python
weather.fillna(weather['평균풍속'].mean(),inplace=True)
print(weather.loc['2012-02-11'])
```

```python
# 1. 다양한 형식의 연, 월, 일 표시 데이터 리스트
d_list = ["01/03/2018", "01-03-2018", "2018-01-05", "2018/01/06"]

# 년도 값을 추출
print(pd.DatetimeIndex(d_list).year)
# 결과: Int64Index([2018, 2018, 2018, 2018], dtype='int64')

# 월 값을 추출
print(pd.DatetimeIndex(d_list).month)
# 결과: Int64Index([1, 1, 1, 1], dtype='int64')

# 일 값을 추출
print(pd.DatetimeIndex(d_list).day)
# 결과: Int64Index([3, 3, 5, 6], dtype='int64')


# 2. 시간 정보가 포함된 데이터 리스트
dt_list = ["01/03/2018 11:12:13", "01-03-2018 11:22:13"]

# 시 값을 추출
print(pd.DatetimeIndex(dt_list).hour)
# 결과: Int64Index([11, 11], dtype='int64')

# 분 값을 추출
print(pd.DatetimeIndex(dt_list).minute)
# 결과: Int64Index([12, 22], dtype='int64')
```
