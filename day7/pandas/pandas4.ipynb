```python
import pandas as pd
# 첫 번째 데이터프레임 df_1 생성
df_1 = pd.DataFrame({
    'A' : ['a10', 'a11', 'a12'],
    'B' : ['b10', 'b24', 'b25'],
    'C' : ['c10', 'c11', 'c12']
}, index = ['가', '나', '다'])

# 두 번째 데이터프레임 df_2 생성
df_2 = pd.DataFrame({
    'B' : ['b23', 'b24', 'b25'],
    'C' : ['c23', 'c24', 'c25'],
    'D' : ['d23', 'd24', 'd25']
}, index = ['다', '라', '마'])
df_3=pd.concat([df_1,df_2])
# print(df_3)
df_4=pd.concat([df_1,df_2],join='inner')
# print(df_4)
df_5=df_1.merge(df_2,how='inner',on='B')
print(df_5)
df_6=df_1.merge(df_2,how='left',on='B')
print(df_6)
df_7=df_1.merge(df_2,how='right',on='B')
print(df_7)
df_8=df_1.merge(df_2,how='outer',on='B')
print(df_8)
```
